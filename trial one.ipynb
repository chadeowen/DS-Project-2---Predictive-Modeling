```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
X = df.drop(columns=['status_group'])
y = df.status_group
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, train_size=0.8, test_size=0.2, random_state=42, shuffle=True)
    
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
majority_class = y_train.mode()
y_pred = np.full(shape=y_test.shape, fill_value=majority_class)
```

```python
# baseline accuracy using majority class

accuracy_score(y_test, y_pred)
```

# First Pass

```python
# first pass will be standard logistic regression with numerical X variables only
# scaled X variables

X = X.select_dtypes('number')
X = X.drop(columns='id')
X = scale(X)
X = pd.DataFrame(X)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
        X, y, train_size=0.8, test_size=0.2, random_state=42, shuffle=True)
    
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
log_reg = LogisticRegression().fit(X_train, y_train)
y_pred = log_reg.predict(X_train)
accuracy_score(y_train, y_pred)
```

```python
log_reg.score(X_train, y_train)
```

```python
# load test data and shape to same as X_train

test = pd.read_csv('/Users/Chad/Downloads/test_features.csv')
test = test.select_dtypes('number')
test_X = test.drop(columns='id')
test_X = scale(test_X)
test_X = pd.DataFrame(test_X)

print(test_X.shape)
test_X.head()
```

```python
# predict for y, test, recombine ids to new y prediction data frame

y_pred = log_reg.predict(test_X)
y_pred = pd.DataFrame(y_pred)
y_pred.head()
```
